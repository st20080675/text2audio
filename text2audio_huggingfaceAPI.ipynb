```python
from transformers import pipeline
import numpy as np
```

```python
pipe = pipeline(model="suno/bark-small")
# output = pipe("Hey it's HuggingFace on the phone!") # this one sounds bad
output = pipe("Hello world. My name is Agnes.")  # sounds ok

audio = output["audio"]
sampling_rate = output["sampling_rate"]
```
